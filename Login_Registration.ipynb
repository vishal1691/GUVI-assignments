```python
import os

def option():
  print("Enter an option:")

  option = (input(" \n option 1. Register \n option 2. Login \n option 3. Forgot Password \n"))
  if option == '1':
    return Register()
  elif option == '2':
    return check_details()
  elif option == '3':
    return Forgot_pass()
  else:
    return 'Enter a valid option'

PASSWORD_FNAME = "/content/sample_data/entries.txt"

def get_existing_users():
  with open(PASSWORD_FNAME, "r") as fp:
    for line in fp.readlines():
      username, password = line.split()
      yield username, password

def is_authorized(username, password):
  return any((user == (username, password) for user in get_existing_users()))

def user_exists(username):
  return any((usr_name == username) for usr_name, password in get_existing_users())

def retrieve(name):
  for username, password in get_existing_users():
    if username == name:
      return "The retrieved password is:"+password

def check_email_password():
  name = email_check()
  password = password_check()
  return name, password

def ask_user_credentials():
  print("Please provide")
  name = str(input("Name:"))
  password = str(input("Password:"))
  return name, password

def check_details():
  name, password = ask_user_credentials()
  if is_authorized(name, password):
    return "Welcome Back, " + name
  if user_exists(name):
    return "Password entered is wrong"
  return "Name not found. Please Sign Up."

def Forgot_pass():
  name, password = ask_user_credentials()
  if user_exists(name):
    return retrieve(name)
  else:
    return 'No user exists. Please Sign Up'

def Register():
  name, password = check_email_password()
  if name == 'Invalid Email ID':
    return 'Enter a valid Email ID'
  if password == 'Invalid Password':
    return 'Enter a valid Password'
  elif user_exists(name):
    return 'User already exists. Please try with a different name'
  else:
    db = open("/content/sample_data/entries.txt","a")
    db.write(name + " " + password + os.linesep)
    print("Registration success!!!")
    db.close()

def email_check():
  email = str(input("Email ID: "))
  X = list(email)
  if '@' in email and '.' in email:
    diff = X.index('.')-X.index("@")
    d,e = 0,0
    if diff > 1:
      d+=1
    if X[0].isalpha():
      e+=1
    if d==1 & e==1:
      return email
    else:
      print("Email ID should start with alphabets")
      return 'Invalid Email ID'
  else:
    print('Email ID should contain @ and .')
    return 'Invalid Email ID'
    
def password_check():
  password = str(input("Password: "))
  special_char = ['$','@','#','%','~','`','!','*','^','&','/','-']
  val = True
  if len(password) < 5 and len(password) > 16:
    print('length of password should be between 5 and 16')
    val = False
  if not any(item.isdigit() for item in password):
    print('Password should have at least one digit')
    val = False
  if not any (item.isupper() for item in password):
    print('Password should have at least one uppercase letter')
    val = False
  if not any (item.islower() for item in password):
    print('Password should have at least one lowercase letter')
    val = False
  if not any (item in special_char for item in password):
    print('Password should have at least one special characters')
    val = False
  if val:
    return password
  else:
    return 'Invalid Password'

option()
```
